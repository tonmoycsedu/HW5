```python
def plot_data(r,s,y_pred,test,y_label):
    
    plot_x = []
    for i in range(r,r+s):
        plot_x.append(i)
        
    plt.plot(plot_x,test,label="original")
    plt.plot(plot_x,y_pred,label=y_label)
    plt.xlabel('Time')
    plt.ylabel('Average request Rate')
    pylab.legend(loc='upper left')   
    plt.show()
```

```python
x = [[1,1],[2,5],[3,3],[4,5],[6,10]]
```

```python
def EWMA(alpha):
    
    training = list(x[:288])
    test = list(x[288:])
    data = training
    
    r= len(training)
    s = len(test)
    
    #print(r,s)
    error = []
    y = []
    y_test= []
    for i in range(r,r+s):
        base = 0
        j = i-1
        k = 0
        #print(data)
        while(j >= 0):
            base = base+ (data[j]*((1-alpha)**k))
            #print(base)
            j = j-1 
            k = k+1
        #print("kdjdkjfkdj")
        #print(base)
        y_pred = alpha*base
        #print(y_pred)
        y.append(y_pred)
        y_test.append(x[i])
        data.append(y_pred)
        
        
        error.append((abs(y_pred-x[i])/x[i])*100)
        
    plot_data(r,s,y,test,"EWMA("+str(alpha)+")")
    
    return sum(error)/len(error)
```

```python
def Seasonal(S):
    training = list(x[:288])
    test = x[288:]
    
    data = training
    r = len(training)
    s = len(test)

    error = []
    y = []
    for i in range(r, r + s):
        y_pred = data[i - S]
        y.append(y_pred)
        data.append(y_pred)

        error.append((abs(y_pred - x[i]) / x[i]) * 100)

    plot_data(r, s, y, test, "Seasonal(" + str(S) + ")")

    return sum(error) / len(error)

```

```python
error_seasonal = Seasonal(288)
print(error_seasonal)
```

```python
def train_lr(p, length,data):
    # print("jhjd")
    training = []
    label = []

    for i in range(length):
        if (i + p < length):
            # print("ghttt")
            training.append([1])
            training[i] = training[i] + list(data[i:i + p])
            # training.append(x[i:i+144])
            label.append(data[i + p])

        else:
            break

    # print(len(label))
    beta = np.matmul(np.linalg.inv(np.matmul(np.transpose(training), training)),
                     np.matmul(np.transpose(training), label))

    return beta


def AR(p):
    training = list(x[:288])
    test = x[288:]

    r = len(training)
    s = len(test)
    data = training
    error = []
    y = []
    for i in range(r, r + s):
        # print(i)
        testx = [1]
        temp = data[i - p:i]
        testx = testx + list(temp)
        # print("ggg")
        beta = train_lr(p, i,data)
        y_pred = np.matmul(testx, beta)
        data.append(y_pred)
        y.append(y_pred)

        error.append((abs(y_pred - x[i]) / x[i]) * 100)

    plot_data(r, s, y, test, "AR(" + str(p) + ")")

    return sum(error) / len(error)

error_ar = AR(3)
print(error_ar)
```
